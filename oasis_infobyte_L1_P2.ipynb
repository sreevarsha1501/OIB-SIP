```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# loading the dataaset
df = pd.read_csv("/content/ifood_df.csv")
print(df.head())
```

```python
# list of all columns
df.columns
```

```python
# Drop missing values (if any)
df.dropna(inplace=True)

# Check for duplicates
df.drop_duplicates(inplace=True)
```

```python
# feature scaling
features = [
    'Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
    'NumStorePurchases', 'NumWebVisitsMonth', 'Age'
]

X = df[features]
```

```python
# applying standard scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# applying k-maens
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# plotting visual
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='MntWines', y='MntMeatProducts', hue='Cluster', palette='Set1')
plt.title("Customer Segments Based on Spending")
plt.show()
```
