# *Exploratory Data Analysis (EDA) on Retail Sales Data*

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loading the datasets
#load menu.csv file
menu_df = pd.read_csv('/content/menu.csv')
#load sales.csv file
sales_df = pd.read_csv('/content/retail_sales_dataset.csv')

#menu data details
menu_df.head()
menu_df.info()
menu_df.describe()

#retail sales details
sales_df.head()
sales_df.info()
sales_df.describe()

#data cleaning
#menu_df data cleaning
menu_df.isnull().sum()
menu_df.dropna(inplace=True)
#sales_df data cleaning
sales_df.isnull().sum()
sales_df.dropna(inplace=True)
```

```python

#plottings
#histogram of calories
#histogram of calories with colormap
#generate a list of colors
#create a colormap
#calculate histogram values
counts, bins = np.histogram(menu_df['Calories'], bins=20)

#create a colormap with 20 distinct colors
colors = plt.cm.viridis(np.linspace(0, 1, len(counts)))

#plot each bar with different color
plt.figure(figsize=(10,6))
for count, x, color in zip(counts, bins, colors):
  plt.bar(x, count, width=np.diff(bins)[0] ,color=color, edgecolor = 'black')
plt.title("Distribution of Colors")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.show()
```

```python
#b) total fat vs calories Scatter plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(menu_df['Total Fat'], menu_df['Calories'], color = 'green')
plt.title("Total Fat vs Calories")
plt.xlabel("Total Fat")
plt.ylabel("Calories")
plt.show()
```

```python
#c) Calories from fat Distribution
#boxplot for calories from fat
plt.figure(figsize=(10,6))
plt.boxplot(menu_df['Calories from Fat'])
plt.title("Calories from Fat")
plt.show()
```

```python
#d) correlations heatmap
#filter from numeric columns only
numeric_menu_df = menu_df.select_dtypes(include=['float64','int64'])
#correlations from correlation matrix
correlation_matrix = numeric_menu_df.corr()
#create a heatmap
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title("Correlations Heatmap")
plt.show()
```

```python
#ii)Supermart grocery sales dataset analysis
#a) sales distribution
#histogram of quantity
#calculate hsitogram values
counts, bins = np.histogram(sales_df['Quantity'], bins=20)

#create a colormap with 20 different colors
colors = plt.cm.viridis(np.linspace(0, 1, len(counts)))
#plot each bar with different color
plt.figure(figsize=(10,6))
for count, x, color in zip(counts, bins, colors):
  plt.bar(x, count, width=np.diff(bins)[0], color=color, edgecolor='black')
plt.title("Distribution of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

```python
#b) profit vs distribution scatter plot
plt.figure(figsize=(10,6))
plt.scatter(sales_df['Age'], sales_df['Quantity'], color='green', marker = '*')
plt.title("Age vs Quantity")
plt.xlabel("Age")
plt.ylabel("Quantity")
plt.show()
```

```python
#c) sales by category bar plot
plt.figure(figsize=(10,6))
sales_by_category = sales_df.groupby('Product Category')['Quantity'].sum()
sns.barplot(x=sales_df['Product Category'], y=sales_df['Quantity'],hue = sales_df['Product Category'], palette = 'Set1')
plt.title("Sales by Category")
plt.xlabel("Product Category")
plt.show()
```
